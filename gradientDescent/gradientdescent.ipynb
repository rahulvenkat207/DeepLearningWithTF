```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
%matplotlib inline
```

```python
df = pd.read_csv("insu_data.csv")
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df[['age','affordibility']],df.bought_insurance,test_size = 0.1)
```

```python
x_train_scaled = x_train.copy()
x_train_scaled['age'] = x_train_scaled['age'] / 100

x_test_scaled = x_test.copy()
x_test_scaled['age'] = x_test_scaled['age'] /100
```

```python
x_train_scaled
```

```python
model = keras.Sequential([
    keras.layers.Dense(1,input_shape=(2,),activation = 'sigmoid',kernel_initializer='ones',bias_initializer='zeros')
])
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
model.fit(x_train_scaled,y_train,epochs = 10000)
```

```python
model.evaluate(x_test_scaled,y_test)
```

```python
def sigmoid(x):
	return 1/(1+np.exp(-x))
```

```python
def prediction_function(age,affordibility):
	weighted_sum = coef[0] * age + coef[1] * affordibility + intercept
	return sigmoid(weighted_sum)

prediction_function(.47,1)
```

```python
def log_loss_function(y_true, y_predicted):
    epsilon = 1e-15

    y_predicted_new = np.maximum(epsilon, y_predicted)
    y_predicted_new = np.minimum(1 - epsilon, y_predicted_new)

    return -np.mean(y_true * np.log(y_predicted_new) + (1 - y_true) * np.log(1 - y_predicted_new))
```

```python
def gradient_descent(age,affordability,y_true,epochs,loss_th):
  w1 = w2 = 1
  bias = 0
  rate = 0.5
  n=len(age)
  for i in range(epochs):
    weighted_sum = w1 * age + w2 * affordability +bias
    y_predicted = sigmoid(weighted_sum)

    loss = log_loss_function(y_true,y_predicted)

    w1d = (1/n) * np.dot(np.transpose(age),(y_predicted - y_true))
    w2d = (1/n) * np.dot(np.transpose(affordability),(y_predicted - y_true))
    bias_d = np.mean(y_predicted - y_true)

    w1 = w1 - rate * w1d
    w2 = w2 - rate * w2d
    bias = bias - rate* bias_d
    if loss <= loss_th:
      break


    print(f"Epoch:{i},w1:{w1},w2:{w2},loss:{loss}")






```
